```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.arange(1,11)
y=x+5
plt.title("Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'*')
plt.show()
```

```python
plt.plot(x,y,'v r')
plt.show()
```

```python
plt.plot(x,(3*x)+y,'v r')
plt.show()
```

```python
x=np.arange(1,11)
y=3*x+5
plt.title("Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.plot(x,y,color="orange",marker='v',markersize='12',label='graph')
plt.plot(x,x+2,marker="x",label='gra')
plt.legend()
plt.show()
```

```python
x=["2000","2001","2002","2003","2004"]
y=[100,159,300,256,567]
z=[10,26,58,110,126]
plt.plot(x,y,color='orange',marker='o',label='popln')
plt.plot(x,z,color='green',marker='v',label='death')
plt.legend()
plt.show()
```

<h1>Prg 2</h1>

```python
x=np.arange(0,3*np.pi,.1)
y=np.sin(x)
plt.title("sine graph")
plt.xlabel("x")
plt.ylabel("sinx")
plt.plot(x,y)
plt.show()
```

```python
plt.title("cosine graph")
plt.xlabel("x")
plt.ylabel("cosx")
plt.plot(x,np.cos(x))
plt.show()
```

```python
plt.title("tangent graph")
plt.xlabel("x")
plt.ylabel("tanx")
plt.plot(x,np.tan(x))
plt.show()
```

```python
x=np.arange(0,3*np.pi,.1)
y=np.sin(x)
z=np.cos(x)
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title('sine wave')
plt.subplot(1,2,2)
plt.plot(x,z)
plt.title('cosine wave')
plt.show()
```

Write a code to plot a line chart to depict the run rate of  a T20 match<br>
Over|runs<br>
5   |45<br>
10  |79<br>
15  |145<br>
20  |234<br>
Also change the thickness of the line<br>

```python
x=[5,10,15,20]
y=[45,79,145,234]
z=[]
for i,j in zip(x,y):
    z.append(j/i)
plt.plot(x,z,color='blue',lw=4,label='runrate',)
plt.xlabel('over')
plt.ylabel('runrate')
plt.title("cricket")
plt.show()
```
