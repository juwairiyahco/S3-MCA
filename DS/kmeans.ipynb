```python
import  numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets  import make_blobs
import matplotlib.pyplot as plt

```

```python
data,_=make_blobs(n_samples=300)
print(data)
```

```python
m=KMeans(n_clusters=5,random_state=42,n_init='auto')
m.fit(data)
labels=m.labels_ #cluster assignmemt for each data point.
set(labels) #set is a function which is used to set the lables for each points.
c=m.cluster_centers_ #it stores the coordinates of the centroids for these 5 clusters.
plt.scatter(data[:,0],data[:,1],c=labels,cmap="viridis")#coloring each points according to the clustering labels
plt.scatter(c[:,0],c[:,1],c='red')
plt.legend(list(labels))
m.predict([[2,-4],[6,6],[6,-2]])



```
